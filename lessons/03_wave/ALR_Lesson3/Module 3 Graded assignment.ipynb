```python
import sympy
sympy.init_printing()

u_max, u_star, rho_max, rho_star, A, B = \
sympy.symbols('u_max u_star rho_max rho_star A B')

eq1 = sympy.Eq( 0, u_max*rho_max*(1 - A*rho_max-B*rho_max**2) )
eq2 = sympy.Eq( 0, u_max*(1 - 2*A*rho_star-3*B*rho_star**2) )
eq3 = sympy.Eq( u_star, u_max*(1 - A*rho_star - B*rho_star**2) )
# use LHS and RHS
eq4 = sympy.Eq(eq2.lhs - 3*eq3.lhs, eq2.rhs - 3*eq3.rhs)
#Two ways to simplify equations
eq4.simplify()
eq4.expand()
# Tell an equation to solve for a variable
rho_sol = sympy.solve(eq4,rho_star)[0]
rho_sol
B_sol = sympy.solve(eq1,B)[0]
B_sol
#Substution
quadA = eq2.subs([(rho_star, rho_sol), (B,B_sol)])
quadA

A_sol = sympy.solve(quadA, A)
A_sol[0]
A_sol[1]

```

```python
#ρ_max=10.0, u_max = 1.0, U_star = 0.7
# Positive result
aval = A_sol[0].evalf(subs={u_star: 1.5, u_max:2.0, rho_max:15.0} )
aval
```

```python
%matplotlib inline
import numpy
import matplotlib.pyplot as plt

from matplotlib import animation
from JSAnimation.IPython_display import display_animation

def u_initial(nx):
    
    u = numpy.ones(nx)
    u[(nx)/2:] = 0
    return u

nx = 81
nt = 70
dx = 4.0/(nx-1)

computeF = lambda u: (u/2)**2

def maccormack(u, nt, dt, dx, eps):
    un = numpy.zeros((nt,len(u)))
    un[:,:] = u.copy()
    ustar = u.copy()
    
    for i in range(1,nt):
        F = computeF(u)
        
        ustar[1:-1] = 1/2*(u[2:] + u[1:-1]) - dt/(2*dx) * (F[2:]-F[1:-1])
        
        Fstar = computeF(ustar)
        
        un[i,1:] = 0.5 * (u[1:]+ustar[1:] - dt/dx * (Fstar[1:] - Fstar[:-1]))
        
        u = un[i].copy()
        
    return un


def animate(data):
    x = numpy.linspace(0,4,nx)
    y = data
    line.set_data(x,y)
    return line,

u = u_initial(nx)
sigma = 1
dt = sigma*dx
eps = 0.25

u = u_initial(nx)
sigma = 1
dt = sigma*dx

un = maccormack(u,nt,dt,dx, eps)

fig = plt.figure();
ax = plt.axes(xlim=(0,4),ylim=(-.5,2));
line, = ax.plot([],[],lw=2);

anim = animation.FuncAnimation(fig, animate, frames=un, interval=50)
display_animation(anim, default_mode='once')
```
